## Rules for Naming Variables

- A variable name can not start with a number, though it can be find anywhere else e.g 15name = oliatan
- A variable name can not contain any special character(s) except underscore name% = olaitan instead namme_ = olaitan
- A variable name can not be join with space, use underscore instead e.g first name = olaitan instead first_name_ = olaitan
- A variable can not be name with python reserved key words e.g print(print)
